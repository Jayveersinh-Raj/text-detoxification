# Raw dataframe from tsv

```python
import pandas as pd
df = pd.read_csv("/content/filtered.tsv", sep="\t")
df
```

```python
df.columns = ["index", "reference",	"translation", "similarity",	"lenght_diff"	, "ref_tox", 	"trn_tox"]
df
```

# Making the train dataset for model training with one-to-one toxic sentence and corresponding detoxified sentences in a dataframe (if toxicity or either reference or translation is above 0.50 and the corresponding other has less than 0.50 toxicity, they both are appended to toxic sentences container, and detoxified container respectively).

```python
non_toxic, toxic = [],[]
for rows, columns in df.iterrows():
  if(columns['ref_tox'] < 0.5 and columns['trn_tox'] > 0.5):
    non_toxic.append(columns['reference'])
    toxic.append(columns['translation'])

  elif(columns['ref_tox'] > 0.5 and columns['trn_tox'] < 0.5):
    toxic.append(columns['reference'])
    non_toxic.append(columns['translation'])
```

```python
len(non_toxic), len(toxic)
```

```python
non_toxic[-1], toxic[-1], non_toxic[-3], toxic[-3]
```

# Making the new dataframe for csv train data from the toxic and detoxified sentences containers (list)

```python
new_df = pd.DataFrame()
new_df["toxic"] = toxic
new_df["detoxified"] = non_toxic
```
